```python
import pandas as pd
import numpy as np
import warnings
import logging
import random
import requests
import sys
import re
from groq import Groq
warnings.filterwarnings('ignore', category=UserWarning)
```

```python
df = pd.read_csv('talents.csv')
df
```

```python

# Get a sample of available job titles to reduce prompt size
# You can adjust the sample size (e.g., 100 or 200) based on testing to stay within the token limit
sample_size = 200
sample_candidates = df['title'].sample(min(sample_size, len(df))).tolist()
```

```python
# Initialize Groq LLM (replace with actual API call)
client= Groq(api_key="<api key>")
```

```python
search_term = "ML Engineer"
```

```python


# Construct prompt
prompt = f"""
Rank the top 5 most similar job titles to "{search_term}" based on relevance.
Available job titles: {sample_candidates}
Provide the results in a numbered list 1 to 5 with 1 being the most relevant. Do not explain.
"""

# Call Groq API
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "system", "content": "You are an AI that ONLY outputs a list of top 5 job titles in strict numbered format. Do NOT explain, think, or reason."},
        {"role": "user", "content": prompt}
    ]
)

# Extract ranked job titles
top_5_jobs = response.choices[0].message.content.split("\n")[:5]

#print(top_5_jobs)
top_5_jobs
```

```python


# Construct prompt
# Construct strict output prompt
prompt = f"""
Available job titles:
{sample_candidates}

Search term: {search_term}

Return ONLY the top 5 most relevant job titles based on the search term.
Respond in this exact format:

1. <job title>
2. <job title>
3. <job title>
4. <job title>
5. <job title>
"""



# Call Groq API
response = client.chat.completions.create(
    model="deepseek-r1-distill-llama-70b",
    messages=[
        {"role": "system", "content": "You are an AI that ONLY outputs a list of top 5 job titles in strict numbered format. Do NOT explain, think, or reason."},
        {"role": "user", "content": prompt}
    ]
)


# Extract ranked job titles
output = response.choices[0].message.content.strip().split("\n")
top_5_jobs = [line for line in output if re.match(r"^\d\.\s", line)][:5]

for job in top_5_jobs:
    print(job)
```
